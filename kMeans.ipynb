```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,random_state=1)
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,6))
ax1.set_title('kmeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title('orignal')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```
